```python
import everyday_pb2

everyday = everyday_pb2.EveryDay()
diary_path = 'diary.txt'
with open(diary_path, 'rb') as f:
    saved_bytes = f.read()
everyday.MergeFromString(saved_bytes)
```

```python
import json
from dateutil.parser import parse

days = []
for oneday in everyday.oneday:
    #date = str(parse(oneday.date))
    date = oneday.date
    text = ""
    for content in oneday.content:
        text += content.date + "\n\n\n\n"
    text.strip()
    type = "kivydiary"
    images = []
    images = json.dumps(images)
    days.append((date, type, text, images))
```

```python
text = ""
for day in days:
    splitor = "\n\n——————————————\n\n" 
    text += splitor + day[0] + "\n\n\n" + day[2] 
print(text)
```

```python
import os
import sqlite3

class Self():
    _root_path = "."
    pass

self = Self()
self.SQL_DATA_FILE = os.path.join(self._root_path, "kivydiary2.db")
if not os.path.exists(self.SQL_DATA_FILE):
    print(f"SQL doesn't exist: {self.SQL_DATA_FILE}")
    print(f"We'll create a new one!")
self._sql_conn = sqlite3.connect(self.SQL_DATA_FILE, check_same_thread=False)
self._sql_cursor = self._sql_conn.cursor()
self._sql_cursor.execute('''CREATE TABLE IF NOT EXISTS thoughts
            (date TEXT, type TEXT, text TEXT, images TEXT)''')

for day in days:
    #print(day)
    self._sql_cursor.execute("INSERT INTO thoughts VALUES (?,?,?,?)", day)
    self._sql_conn.commit()
```

```python
#https://docs.python.org/3/library/sqlite3.html
#https://pynative.com/python-sqlite-blob-insert-and-retrieve-digital-data/
import sqlite3
conn = sqlite3.connect('example.db')
c = conn.cursor()
```

```python
# Create table
c.execute('''CREATE TABLE IF NOT EXISTS thoughts 
             (date TEXT, data BLOB)''')

# Insert a row of data
c.execute("INSERT INTO thoughts VALUES (?,?)", ('2006-01-05',b'y ingshaoxo said'))

# Larger example that inserts many records at a time
purchases = [
    ('2006-03-28', b'hi'),
    ('2006-04-05', b"everyone"),
]
c.executemany('INSERT INTO thoughts VALUES (?,?)', purchases)

# Save the changes
conn.commit()

# We can also close the connection if we are done with it.
# Just be sure any changes have been committed or they will be lost.
#conn.close()
```

```python
for row in c.execute('SELECT * FROM thoughts ORDER BY date'):
    print(row)
```

```python
target = ('2006-03-28',)
c.execute('SELECT * FROM thoughts WHERE date=?', target)
print(c.fetchone())
```

```python
#get the count of tables with the name
c.execute(''' SELECT * FROM thoughts WHERE date='2006-03-28' and date='2006-03-28' ''')

r = c.fetchall()
print(r)
print(len(r))

#if the count is 1, then table exists
if r:
    print('Table exists.')
```
